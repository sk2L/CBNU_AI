```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('titanic.csv')
```

```python
# DataFrame 데이터 탐색
# 1. 생존자 중 몇 명인지 계산하세요.
survived_count = df['Survived'].sum()
print("생존자 수:", survived_count)
```

```python
# 2. 항구 "S"에서 탑승한 승객 수를 계산하세요.
embarked_S_count = (df['Embarked'] == 'S').sum()
print("항구 'S'에서 탑승한 승객 수:", embarked_S_count)
```

```python
# 3. 남성 승객 수를 계산하세요.
male_passengers = (df['Sex'] == 'male').sum()
print("남성 승객 수:", male_passengers)
```

```python
# 4.3등실에 탑승하고 지불한 요금이 40.000 미만인 승객 수를 계산하세요.
third_class_low_fare = ((df['Pclass'] == 3) & (df['Fare'] < 40.000)).sum()
print("3등실에 탑승하고 40.000 미만 요금 지불한 승객 수:", third_class_low_fare)
```

```python
# 5. 승객 ID가 674인 승객의 이름을 찾아보세요.
passenger_name = df.loc[df['PassengerId'] == 674, 'Name'].values[0]
print("승객 ID가 674인 승객의 이름:", passenger_name)
```

```python
# 6. 승객들의 평균 나이를 계산하세요.
average_age = df['Age'].mean()
print("승객들의 평균 나이:", average_age)
```

```python
# 7. 생존자와 사망자의 수를 각각 계산하세요.
survived_count = df['Survived'].sum()
not_survived_count = len(df) - survived_count
print("생존자 수:", survived_count)
print("사망자 수:", not_survived_count)
```

```python
# 8.남성과 여성 승객 수를 계산하세요.
male_count = (df['Sex'] == 'male').sum()
female_count = (df['Sex'] == 'female').sum()
print("남성 승객 수:", male_count)
print("여성 승객 수:", female_count)
```

```python
# 9. 각 성별에 따른 생존자와 사망자의 수를 계산하세요.[DataFrame. groupby() 함수 이용]
gender_survival = df.groupby('Sex')['Survived'].value_counts().unstack()
print("각 성별에 따른 생존자와 사망자 수:")
print(gender_survival)
```

```python
# 10. 각 등급별로 생존한 사람과 사망한 사람의 평균 지불 요금을 계산하세요[(DataFrame.pivot_table() 함수 이용)
fare_by_pclass = df.pivot_table(index='Pclass', columns='Survived', values='Fare', aggfunc='mean')
print("각 등급별로 생존한 사람과 사망한 사람의 평균 지불 요금:")
print(fare_by_pclass)
```

```python
# DataFrame 데이터 시각화
# 1. 탑승자의 생존 여부에 대한 비율을 계산하고, 이 비율을 막대 그래프로 시각화.
survival_ratio = df['Survived'].value_counts(normalize=True)
survival_ratio.plot(kind='bar', rot=0)
plt.title("탑승자 생존 여부 비율")
plt.xlabel("생존 여부")
plt.ylabel("비율")
plt.show()
```

```python
# 2.나이 열의 분포를 20개의 구간(bin)으로 나누어 히스토그램으로 시각화
df['Age'].hist(bins=20)
plt.title("나이 분포")
plt.xlabel("나이")
plt.ylabel("빈도")
plt.show()
```

```python
# 3. 나이와 요금 간의 산점도 시각화
plt.scatter(df['Age'], df['Fare'])
plt.title("나이와 요금 간의 산점도")
plt.xlabel("나이")
plt.ylabel("요금")
plt.show()
```

```python
# 4. 객실 등급(Pclass)별로 생존자와 사망자 수를 막대그래프로 시각화.
pclass_survival = df.groupby('Pclass')['Survived'].value_counts().unstack()
pclass_survival.plot(kind='bar', rot=0)
plt.title("객실 등급별 생존자와 사망자 수")
plt.xlabel("객실 등급")
plt.ylabel("인원 수")
plt.show()
```
